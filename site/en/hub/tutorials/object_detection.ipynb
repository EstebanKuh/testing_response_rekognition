```

# Descargar la imagen desde la URL
response = requests.get(url_imagen)
img_data = response.content

# Cargar la imagen en Pillow
img = Image.open(io.BytesIO(img_data))
draw = ImageDraw.Draw(img)

# Dibujar los cuadros delimitantes y las etiquetas desde el resultado de Lambda
for label in resultado_lambda["etiquetas"]:
    bounding_box = label["posiciones"]
    width, height = img.size
    left = int(bounding_box["Left"] * width)
    top = int(bounding_box["Top"] * height)
    right = int((bounding_box["Left"] + bounding_box["Width"]) * width)
    bottom = int((bounding_box["Top"] + bounding_box["Height"]) * height)

    # Dibujar el rectngulo (cuadro delimitante)
    draw.rectangle([left, top, right, bottom], outline="red", width=3)

    # Dibujar la etiqueta encima del cuadro delimitante
    label_text = f"{label['etiqueta']} ({label['confianza']:.2f}%)"
    font = ImageFont.load_default()
    draw.text((left, top - 10), label_text, fill="white", font=font)

# Mostrar la imagen con los cuadros y etiquetas
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')  # Ocultar los ejes
plt.show()
```
